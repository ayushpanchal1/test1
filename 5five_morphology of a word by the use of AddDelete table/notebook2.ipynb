```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import words
import pandas as pd
```

```python
nltk.download('words')
nltk.download('omw-1.4')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
eng_w = set(words.words())
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
def modword(word, rempre, remsuff):
    if rempre and word.startswith(rempre):
      word = word[len(rempre):]
    if remsuff and word.endswith(remsuff):
        word = word[: -len(remsuff)]
    return word
```

```python
prefixes = ['un','bi','diss','miss','pre','post']
suff = ["ing", 'ed', 's', 'ly', 'er', 'est', 'ment', 'tion', 'able', 'ible', 'ness', 'iness']

data = []

word_list = ["interesting", "approachable", "uncomfortable", "multiplication", "attachment", "representation", "appreciation", "generation", "running", "cars", "happily","unhappiness"]
```

```python
for word in word_list:
  original_word = word
  suffix_removed = ""
  prefix_removed = ""
  stemmed_word = ""
  lemmatized_word = ""
  affixes_removed = ""

  for suf in suff:
      if word.endswith(suf):
          suffix_removed = suf
          break
      else:
          suffix_removed = word
  for p in prefixes:
      if word.startswith(p):
          prefix_removed = p
          break

  affixes_removed = modword(word, prefix_removed, suffix_removed)
  stemmed_word = stemmer.stem(word)
  lemmatized_word = lemmatizer.lemmatize(word)

  data.append([original_word, prefix_removed, suffix_removed, affixes_removed, stemmed_word, lemmatized_word])

df = pd.DataFrame(data, columns=["Original Word", "Prefix Removed", "Suffix Removed", "Without Affixes", "Stemmed Word", "Lemmatized Word"])
df
```
